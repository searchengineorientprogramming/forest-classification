```python
import matplotlib.pyplot as plt
import numpy as np
import os
from obspy import read
from obspy import UTCDateTime
import pandas as pd
from scipy import stats
from scipy.fft import rfft, rfftfreq
```

### Split the raw signal into 10-minute section at 12 time windows of the day, do FFT on each of the 12 time window and save to .parquet format under the folder of `raw_FFT`

```python
tree = "Forest\\DP1"
nontree = "Non-forest\\DP1"

data_row = []

for station in [item for item in os.listdir(tree)]:
    print(f'working on {station}')
    for file in os.listdir(tree + '\\' + station):
        print(f'working on {file}')
        path = tree + '\\' + station + '\\' + file

        stream = read(path)
        start_date = min(trace.stats.starttime.date for trace in stream)
        start_time = UTCDateTime(f'{start_date}T22:00:00')
        end_time = start_time + 600
        cropped_stream = stream.trim(starttime=start_time, endtime=end_time)
        amp = np.abs(rfft(cropped_stream[0]))
        freq = rfftfreq(len(cropped_stream[0]), 1/cropped_stream[0].meta['sampling_rate'])
        mask = (freq > 10) & (freq < 100)
        filtered_freq = freq[mask]
        filtered_amp = amp[mask]
        data_row.append(list(filtered_amp) + ['Tree'])

for station in [item for item in os.listdir(nontree)]:
    print(f'working on {station}')
    for file in os.listdir(nontree + '\\' + station):
        print(f'working on {file}')
        path = nontree + '\\' + station + '\\' + file

        stream = read(path)
        start_date = min(trace.stats.starttime.date for trace in stream)
        start_time = UTCDateTime(f'{start_date}T22:00:00')
        end_time = start_time + 600
        cropped_stream = stream.trim(starttime=start_time, endtime=end_time)
        amp = np.abs(rfft(cropped_stream[0]))
        freq = rfftfreq(len(cropped_stream[0]), 1/cropped_stream[0].meta['sampling_rate'])
        mask = (freq > 10) & (freq < 100)
        filtered_freq = freq[mask]
        filtered_amp = amp[mask]
        data_row.append(list(filtered_amp) + ['Nontree'])

df = pd.DataFrame(data_row, columns = [f'M{i+1}' for i in range(len(filtered_amp))]+['result'])

df.to_parquet('raw_FFT/raw_data_2200_11.parquet')
```

### Prepare dataframe: read the FFT spectrum from .parquet file in `raw_FFT`, perform min_max_normalization and evenly pick 100 amplitude as training features, then save to file `raw_data_fft_DP1.parquet`

```python
def min_max_scale_row(row):
    min_val = row.min()
    max_val = row.max()
    return (row - min_val) / (max_val - min_val)

FFT_data_files = ["raw_data_2200_11.parquet", "raw_data_2220_11.parquet", "raw_data_2240_11.parquet", "raw_data_2300_11.parquet", "raw_data_2320_11.parquet", "raw_data_2340_11.parquet",
                  "raw_data_0600_11.parquet", "raw_data_0620_11.parquet", "raw_data_0640_11.parquet", "raw_data_0700_11.parquet", "raw_data_0720_11.parquet", "raw_data_0740_11.parquet"]
all_data = pd.concat([pd.read_parquet("raw_FFT"+"/"+file) for file in FFT_data_files], axis=0, ignore_index=True)

all_data.columns = list(filtered_freq) + ['results']

X = all_data.drop('results', axis = 1)
X = X.apply(min_max_scale_row, axis=1)
X = X.loc[:, X.columns >= 10]
num_columns_to_pick = 100; step = X.shape[1] // num_columns_to_pick; selected_columns_indices = [i * step for i in range(num_columns_to_pick)]
X = X.iloc[:, selected_columns_indices]
y = all_data.results

df = pd.concat([X, y], axis=1)
df.to_parquet('raw_data_fft_DP1.parquet')
```
