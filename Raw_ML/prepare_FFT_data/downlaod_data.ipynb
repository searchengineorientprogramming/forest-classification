```python
import obspy 
from obspy.clients.fdsn.mass_downloader import RectangularDomain, Restrictions, MassDownloader,CircularDomain
import numpy as np
from tqdm import tqdm

domain = RectangularDomain(minlatitude=62,maxlatitude=64,minlongitude=-146,maxlongitude=-145)
# domain = CircularDomain(latitude=78.59, longitude=-57.2, minradius=0.0, maxradius=0.5)

years = [2016]
months = [4,5]
days = np.arange(1,31)
net = '9M'
sta = [247,260]   # please select the station name
com = 'DP2'    # please select the component: 'DP1','DP2','DPZ'


for st in tqdm(sta):
    for yr in years:
        for mo in months:
            for d in days:
                restrictions = Restrictions(
                    starttime = obspy.UTCDateTime(yr,mo,d,0,0,0),
                    endtime = obspy.UTCDateTime(yr,mo,d,23,59,59),
                    # chunklength_in_sec=86400,
                    network = net,
                    station = st,
                    # location="00",
                    channel = com,
                    reject_channels_with_gaps = False,
                    minimum_length = 0.1,
                    minimum_interstation_distance_in_m=0.0
                    )

                mdl = MassDownloader(providers=['IRIS'])
                mdl.download(domain, restrictions, mseed_storage='9M_raw_changed_dataset/tree - Copy/%s/%s'%(com,st),stationxml_storage='station/')
```
